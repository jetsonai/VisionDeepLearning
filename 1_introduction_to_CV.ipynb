## 4. Moving Average - ì—¬ëŸ¬ window í¬ê¸° ë¹„êµ

```

t = np.linspace(0,20,500)
signal = np.sin(t) + np.random.randn(500)*0.3

plt.figure(figsize=(12,8))
plt.plot(signal,label="signal")

for w in [3,7,15,30]:
    kernel = np.ones(w)/w
    ma = np.convolve(signal,kernel,'same')
    plt.plot(ma,label=f"window={w}")

plt.legend(); plt.show()
```

## 5. Edge Detection (Sobel, Prewitt, Laplacian, Canny)

```

img = cv2.imread("lena.png",0)

sobel = cv2.Sobel(img, cv2.CV_64F, 1,1)
lap = cv2.Laplacian(img, cv2.CV_64F)
canny = cv2.Canny(img,100,200)

# Prewitt manually
kernelx = np.array([[1,0,-1],[1,0,-1],[1,0,-1]])
kernely = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
prew = cv2.filter2D(img,-1,kernelx) + cv2.filter2D(img,-1,kernely)

plt.figure(figsize=(12,8))
titles=["Sobel","Prewitt","Laplacian","Canny"]
for i,res in enumerate([sobel,prew,lap,canny]):
    plt.subplot(2,2,i+1); plt.imshow(res,cmap='gray'); plt.title(titles[i])
plt.show()
```

## 6. Sharpen í•„í„° ì—¬ëŸ¬ ì¢…ë¥˜ ë¹„êµ

```

kernels=[
    np.array([[0,-1,0],[-1,5,-1],[0,-1,0]]),
    np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]]),
    np.array([[1,-2,1],[-2,5,-2],[1,-2,1]])
]
plt.figure(figsize=(12,8))
for i,k in enumerate(kernels):
    out = cv2.filter2D(img, -1, k)
    plt.subplot(1,3,i+1); plt.imshow(out,cmap='gray'); plt.title(f"Sharpen {i+1}")
plt.show()
```

## 7. Box Blur (Kernel size ë³€í™”)

```

plt.figure(figsize=(12,8))
for i,k in enumerate([3,5,9,15]):
    kernel = np.ones((k,k))/k**2
    out = cv2.filter2D(img,-1,kernel)
    plt.subplot(2,2,i+1); plt.imshow(out,cmap='gray'); plt.title(f"Box blur {k}x{k}")
plt.show()
```

## 8. Gaussian Blur (sigma ë³€í™”)

```

plt.figure(figsize=(12,8))
sigmas=[0.5,1,2,4]
for i,s in enumerate(sigmas):
    g = cv2.GaussianBlur(img,(7,7),s)
    plt.subplot(2,2,i+1); plt.imshow(g,cmap='gray'); plt.title(f"Gaussian Ïƒ={s}")
plt.show()
```
